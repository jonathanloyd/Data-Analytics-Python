```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""Problem 1"""

# Make sure the csv file is in the same folder as this notebook
df_init = pd.read_csv('MetObjects_Subset.csv')
df_init
```

```python
"""Problem 2"""

# Deep copy df_init to df_prob2
df_prob2 = df_init.copy()

# Get the number of objects in the dataframe to help determine 50%
number_of_objects = len(df_init.index)

# Get the names of all the columns with at least 50% missing values into a list
drop_columns = list()
for (column_name, column_data) in df_init.iteritems():
    if column_data.isna().sum()/number_of_objects >= .50:
        drop_columns.append(column_name)

# Drop all of the columns with at least 50% missing values and display the dataframe
df_prob2 = df_prob2.drop(columns=drop_columns)
df_prob2
```

```python
"""Problem 3"""

# Get the 10 most common values for “Object Name”
frequency_series = df_prob2["Object Name"].value_counts()
frequency_series = frequency_series.sort_values(ascending=False)
top_10_obj_names = list(frequency_series[0:10].index)
# print(top_10_obj_names)

# Deep copy the dataframe from problem 2 and then make a new dataframe displaying only the items with the top 10 object names
df_prob3 = df_prob2.copy()
df_prob3 = df_prob3[df_prob3["Object Name"].isin(top_10_obj_names) == True]
df_prob3
```

```python
"""Problem 4"""

# Get the number of objects in the dataframe to help with percentages
number_of_objects_4 = len(df_prob3.index)

# Get the Percentage of items from the US, Mexico, Canada, or "Other"
number_us = (df_prob3["Country"] == 'United States').sum()
percent_us = number_us / number_of_objects_4 * 100
number_mex = (df_prob3["Country"] == 'Mexico').sum()
percent_mex = number_mex / number_of_objects_4 * 100
number_can = (df_prob3["Country"] == 'Canada').sum()
percent_can = number_can / number_of_objects_4 * 100
number_other = number_of_objects_4 - number_us - number_mex - number_can
percent_other = number_other / number_of_objects_4 * 100

# Store these results in a dictionary
dict_prob4 = {"United States": percent_us, "Canada": percent_can, 
                "Mexico": percent_mex, "Other": percent_other}
print(dict_prob4)
```

```python
"""Problem 5"""

# Initialize the dictionary
dict_prob5 = {"1900s": 0, "1910s": 0, "1920s": 0, "1930s": 0, "1940s": 0, "1950s": 0, "1960s": 0, "1970s": 0, "1980s": 0, "1990s": 0}


# Check if the date is in the 1900s and then check which decade
# Increment the value for the respective key based on the decade of the item
for date in df_prob3["Object End Date"]:
    if date >= 1900 and date <= 1999:
        if date >= 1900 and date <= 1909:
            dict_prob5["1900s"] += 1
        elif date >= 1910 and date <= 1919:
            dict_prob5["1910s"] += 1
        elif date >= 1920 and date <= 1929:
            dict_prob5["1920s"] += 1
        elif date >= 1930 and date <= 1939:
            dict_prob5["1930s"] += 1
        elif date >= 1940 and date <= 1949:
            dict_prob5["1940s"] += 1
        elif date >= 1950 and date <= 1959:
            dict_prob5["1950s"] += 1
        elif date >= 1960 and date <= 1969:
            dict_prob5["1960s"] += 1
        elif date >= 1970 and date <= 1979:
            dict_prob5["1970s"] += 1
        elif date >= 1980 and date <= 1989:
            dict_prob5["1980s"] += 1
        elif date >= 1990 and date <= 1999:
            dict_prob5["1990s"] += 1
print(dict_prob5)
```

```python
"""Problem 6"""

# Get the key names and values into clean variables for problem 4 dict
prob4_names = list(dict_prob4.keys())
prob4_values = list(dict_prob4.values())

# Plot the bar graph for problem 4, and add information regarding the x axis, y axis, and the title
plt.bar(range(len(dict_prob4)), prob4_values, tick_label=prob4_names)
plt.xlabel("Country")
plt.ylabel("Percent of Objects")
plt.title("Percent of Objects from Select Countries")
plt.show()

# Clear the figure in between plotting
plt.clf()

# Get the key names and values into clean variables for problem 5 dict
prob5_names = list(dict_prob5.keys())
prob5_values = list(dict_prob5.values())

# Plot the bar graph for problem 5, and add information regarding the x axis, y axis, and the title
plt.bar(range(len(dict_prob5)), prob5_values, tick_label=prob5_names)
plt.xlabel("Object End Date Decade")
plt.ylabel("Number of Occurrences")
plt.title("Number of Occurrences for Each Decade in Object End Date")
plt.show()
```
